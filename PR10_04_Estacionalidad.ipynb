```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
!wget "https://raw.githubusercontent.com/Dr-Carlos-Villasenor/PatternRecognition/main/Dataset/daily-min-temperatures.csv"
data = pd.read_csv('daily-min-temperatures.csv')
print(data.head())
```

```
x = np.asanyarray(data[['Temp']])
```

```
x.shape
```

```
dx = x[:-1] - x[1:]
```

```
fig = plt.figure(layout='constrained', figsize=(27,9))
plt.suptitle('Estacionalidad de una función', fontsize=20)

plt.subplot(2, 1, 1)
plt.title('Original')
plt.plot(x)

plt.subplot(2, 1, 2)
plt.title('Diferenciada')
plt.plot(dx)
```
