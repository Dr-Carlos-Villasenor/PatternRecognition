## Frecuencia en series de tiempo


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import acf, pacf
```

```
!wget "https://raw.githubusercontent.com/Dr-Carlos-Villasenor/PatternRecognition/main/Dataset/daily-min-temperatures.csv"
data = pd.read_csv('daily-min-temperatures.csv')
print(data.head())
```

```
x_day = np.asanyarray(data[['Temp']])
```

```
x_week = []
for i in range(0, x_day.shape[0], 7):
  x_week.append(np.mean(x_day[i:i+7]))
x_week = np.array(x_week)
```

```
x_month = []
for i in range(0, x_day.shape[0], 30):
  x_month.append(np.mean(x_day[i:i+30]))
x_month = np.array(x_month)
```

```
x_quarter = []
for i in range(0, x_day.shape[0], 91):
  x_quarter.append(np.mean(x_day[i:i+30]))
x_quarter = np.array(x_quarter)
```

```
x_bianual = []
for i in range(0, x_day.shape[0], 182):
  x_bianual.append(np.mean(x_day[i:i+30]))
x_bianual = np.array(x_bianual)
```

```
x_anual = []
for i in range(0, x_day.shape[0], 365):
  x_anual.append(np.mean(x_day[i:i+30]))
x_anual = np.array(x_anual)
```

```
fig = plt.figure(layout='constrained', figsize=(27,9))
plt.suptitle('Frequency in Time series', fontsize=20)

plt.subplot(6, 1, 1)
plt.title('Daily')
plt.plot(x_day)

plt.subplot(6, 1, 2)
plt.title('Weekly')
plt.plot(x_week)

plt.subplot(6, 1, 3)
plt.title('Monthly')
plt.plot(x_month)


plt.subplot(6, 1, 4)
plt.title('Quarterly')
plt.plot(x_quarter)

plt.subplot(6, 1, 5)
plt.title('Bianual')
plt.plot(x_bianual)

plt.subplot(6, 1, 6)
plt.title('Anual')
plt.plot(x_anual)
```
